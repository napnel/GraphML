```python
import torch
import numpy as np
import matplotlib.pyplot as plt

import torch_geometric.datasets as datasets
```

```python
from pathlib import Path
DATA_PATH = Path().cwd().parent / "datasets"
```

### Geometric Shapes  
Synthetic dataset of various geometric shapes like cubes, spheres or pyramids.

```python
dataset = datasets.GeometricShapes(root=DATA_PATH / "Geometric Shapes")
```

```python
print("The number of Graph:", len(dataset))
print("The number of classes:", dataset.num_classes)
print("Nodes and features:", dataset.num_node_features)
print("Edges and features:", dataset.num_edge_features)
```

### 3D ShapeNets: A deep Representation for Volumetric Shapes

```python
dataset = datasets.ModelNet(root=DATA_PATH / "ModelNet")
print("The number of Graph:", len(dataset))
print("The number of classes:", dataset.num_classes)
print("Nodes and features:", dataset.num_node_features)
print("Edges and features:", dataset.num_edge_features)
```

### A Scalable Active Framework for Region Annotation in 3D Shape Collections

```python
dataset = datasets.ShapeNet(root=DATA_PATH / "ShapeNet")
print("The number of Graph:", len(dataset))
print("The number of classes:", dataset.num_classes)
print("Nodes and features:", dataset.num_node_features)
print("Edges and features:", dataset.num_edge_features)
```

```python
data = dataset[1]
points = data.pos.numpy()
# x, y, z = np.meshgrid(points[0], points[1], points[2])
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(points[:, 0], points[:, 1], points[:, 2])
```
